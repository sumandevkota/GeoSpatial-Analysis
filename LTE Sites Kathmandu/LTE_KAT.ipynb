```python
#Importing necessary Libraries

import pandas as pd
import folium
import geopandas as gpd
from folium import Marker
```

```python
# reading an excel file that contains approximate lat long of a LTE sites in Kathmandu
lte = pd.read_excel("KAT.xlsx")
df = pd.DataFrame(lte) # converting to a pandas data frame

# The data that I have here is the approximate data of the LTE sites in kathmandu
# I don't own the data and I am not sure if the datas are accurate
# The sole purpose of this database it to learn  Marker function on Map
```

```python
df.head().append(df.tail()) # a glance of a data top and bottom five
```

```python
# converting pandas data frame to GeoDataFrame
data = gpd.GeoDataFrame(df,geometry=gpd.points_from_xy(df.Long,df.Lat)) 
```

```python
# checking coordinate reference system used
# when we convert data frame to a geodataframe we initilize CRS to epsg:4326
print(data.crs)
```

```python
# Initilizing CRS to EPSG:4326
data.crs = {'init':'epsg:4326'}
# Our purpose of adding marking using Latitude and Longitude will work even if we don't initilize CRS 
# If we have to work with another GeoPanda Dataframe both of them must have same CRS 
# Look more about Coordinate reference System in internet
```

```python
# conforming the initilized CRS system
print(data.crs)
```

```python
# Having a glance of a geo data sets top and bottom five rows
data.head().append(data.tail())
```

```python
#Creating a Map using latitude and longitude of City Kathmandu
Map = folium.Map(location = [27.7172,85.3240],titles = 'openstreeetmap',zoom_start = 12)
Map
```

```python
# Iterating the data set by each row and plotting the Marker using Latitude and Longitude and topup name as a Name

for idx,row in data.iterrows():
    Marker([row['Lat'],row['Long']],popup = row ['Name']).add_to(Map) # adding Each Marker to a Map
    
Map    
```
